```python
#Construction
df=pd.DataFrame([])

st = 1
ub = 5
lb = 0.25
no = 100


    
li = [list(np.linspace(st,ub,no))
        ,list(np.linspace(st,lb,no))
        ,list(np.linspace(st,ub/2,no)*np.linspace(st,ub/2,no))
        ,list(np.linspace(st,lb/2,no)*np.linspace(st,lb/2,no))
     ]


df=pd.DataFrame(li, index=["linear_rising", "linear_falling", "exp_rising","exp_falling"]).T
df.to_csv("data-sources/scenarios.csv",index=False)
```
